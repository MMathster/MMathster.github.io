# Experimental Geometry: Triangle Centers and Their Loci

This notebook explores several classical triangle centers and visualizes the loci traced by these centers when vertex C moves along a circle.

Centers explored include:
- Centroid (X₂)
- Circumcenter (X₃)
- Orthocenter (X₄)
- Nine-point center (X₅)

Interactive controls allow the user to observe how these centers evolve as the triangle deforms.

# Section A - Geometric Setup

## Section A — Geometric Setup

Let $\triangle ABC$ be a right triangle inscribed in the unit circle.

We fix the diameter endpoints

$$A = (1,0), \qquad  
B = (-1,0)$$

and allow the third vertex to move along the circumcircle

$$C = (\cos\theta, \sin \theta), \qquad  \theta \in [0,2π)$$

By Thales's theorem, $\angle C = \frac{\pi}{2}$ for all $\theta$. The side lengths are

$$a = |BC|, \qquad b = |AC|, \qquad c = |AB| = 2$$

Direct computation gives

$$a = \sqrt{2 + 2\cos\theta}, \qquad b = \sqrt{2 − 2\cos\theta}$$

Using half-angle identities we obtain the simpler expressions

$$a = 2 \sin\left(\dfrac{\theta}{2}\right), \qquad b = 2 \cos\left(\dfrac{\theta}{2}\right), \qquad  c = 2$$

when restricting to the upper semicircle $\theta \in [0,\pi)$.

```python
import sympy as sp
from IPython.display import display

theta = sp.symbols('theta', real=True)

A = sp.Point(1,0)
B = sp.Point(-1,0)
C = sp.Point(sp.cos(theta), sp.sin(theta))

# side lengths
a = B.distance(C)
b = A.distance(C)
c = A.distance(B)

display(sp.simplify(a))
display(sp.simplify(b))
display(sp.simplify(c))
```

Given that the vertices $A$ and $B$ are fixed at $(\pm 1, 0)$, the triangle is always a right triangle by Thales's Theorem. How does this specific constraint simplify the barycentric coordinates $(f : g : h)$ compared to a general scalene triangle?

## Section B — Barycentric Coordinates

Triangle centers are often described using barycentric coordinates

$$(f : g : h)$$

which correspond to the Cartesian point

$$X = \dfrac{fA + gB + hC}{f + g + h}$$

In the present configuration

$$A = (1,0), \qquad B = (-1,0), \qquad C = (\cos\theta, \sin\theta)$$

so the resulting Cartesian coordinates are

\begin{align*}
x =& \dfrac{f - g + h\cos\theta}{f + g + h}\\
y =& \dfrac{h\sin\theta}{f + g + h}
\end{align*}

This formula allows us to compute the parametric curve traced by any triangle center.

```python
def barycentric_to_cartesian(f,g,h,A,B,C):

    denom = sp.simplify(f+g+h)

    x = sp.simplify((f*A.x + g*B.x + h*C.x)/denom)
    y = sp.simplify((f*A.y + g*B.y + h*C.y)/denom)

    return x,y
```

As an example, let's try the centroid denoted as $X_2$ whose barycentric coordinates are $(1:1:1)$.

```python
f,g,h = sp.Integer(1),sp.Integer(1),sp.Integer(1)

x,y = barycentric_to_cartesian(f,g,h,A,B,C)

display(x)
display(y)
```

# Section C - Automatic Parametric Curve Extraction

For any center,

$$
    X_n(\theta) = \lambda_n(\theta)\left(\cos\theta, \sin\theta\right)
$$

which comes from the fact of that in this construction
- $O = (0,0)$;
- $H = \left(\cos\theta,\sin\theta\right)$;

So that for any center $X_n$ lying on the Euler line,

$$
    X_n = O + \lambda_n(H - O) = \left(\lambda \cos\theta, \lambda\sin\theta\right)
$$

Thus, the locus of any Euler-line center with constant $\lambda$ is a circle of radius $|\lambda|$.

```python
def center_curve(f,g,h):

    x,y = barycentric_to_cartesian(f,g,h,A,B,C)

    x = sp.simplify(x)
    y = sp.simplify(y)

    return x,y
```

There can be triangle centers that lie along the circumcircle. For instance, the barycentric coordinates of $X_4$ is

$$
    (\tan\alpha: \tan\beta: \tan\gamma) \equiv (0:0:1)
$$

where $\tan\gamma = \infty$. Hence, coordinates of $X_4$ are exactly those of $C$.

# Section D - Interactive Visualizations

The simplest case is that the center $X_n$ lies on the Euler line. Let's run the following code for circumcenter $X_3$, centroid $X_2$, nine-point center $X_5$ and orthocenter $X_4$, which all lie on the Euler line. Consequently, the locus of $X_n$ that lies on the Euler line admits the polar parametrization

$$
    r(\theta) = |\lambda_n(\theta)|
$$

```python
from IPython.display import display
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

centers = {
    "Circumcenter (X3)": 0,
    "Centroid (X2)": 1/3,
    "Nine-Point (X5)": 1/2,
    "Orthocenter (X4)": 1
}

def plot_center(name):

    lam = centers.get(name)
    if lam is None:
        return

    theta = np.linspace(0,2*np.pi,500)

    x = lam*np.cos(theta)
    y = lam*np.sin(theta)

    fig,ax = plt.subplots(figsize=(5,5))

    ax.plot(np.cos(theta),np.sin(theta),'--',alpha=0.3,label="Circumcircle")
    ax.plot(x,y,label=name)

    ax.set_aspect("equal")
    ax.set_xlim(-1.2,1.2)
    ax.set_ylim(-1.2,1.2)

    ax.legend()
    plt.show()
```

The following is the combined setup of four mentioned Kimberling centers in the Thales configuration. When $\lambda_n(\theta)$ is a constant, the triangle center lies on the Euler line.

```python
def plot_geometry(theta=1.05):
    # 1. Geometry Setup
    A = np.array([1, 0])
    B = np.array([-1, 0])
    C = np.array([np.cos(theta), np.sin(theta)])

    # Kimberling Centers (Thales Configuration)
    O = np.array([0, 0])           # Circumcenter (X3)
    H = C                          # Orthocenter (X4) - identical to C here
    G = (A + B + C) / 3            # Centroid (X2)
    N = (O + H) / 2                # Nine-point Center (X5)

    # 2. Plotting
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Precompute the circles for the Loci
    t = np.linspace(0, 2*np.pi, 500)
    
    # --- VISIBLE LOCI ---
    # We use 'zorder=1' to keep these behind the triangle and points
    # Orthocenter Locus (Radius 1)
    ax.plot(np.cos(t), np.sin(t), color='red', lw=3, alpha=0.15, label="Orthocenter Locus (X4)")
    # Nine-point Locus (Radius 1/2)
    ax.plot(0.5*np.cos(t), 0.5*np.sin(t), color='magenta', lw=3, alpha=0.25, label="Nine-point Locus (X5)")
    # Centroid Locus (Radius 1/3)
    ax.plot((1/3)*np.cos(t), (1/3)*np.sin(t), color='green', lw=3, alpha=0.35, label="Centroid Locus (X2)")

    # --- TRIANGLE & EULER LINE ---
    ax.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], color='navy', lw=2, zorder=2)
    ax.plot([O[0], H[0]], [O[1], H[1]], color='orange', linestyle='--', lw=2, label="Euler Line", zorder=3)

    # --- POINTS & LABELS ---
    # Separate C and H labels specifically
    ax.scatter(*C, color='red', s=80, edgecolors='white', zorder=5)
    ax.text(C[0]+0.05, C[1]+0.05, "C (Vertex)", color='navy', fontweight='bold')
    ax.text(C[0]+0.05, C[1]-0.08, "H (Orthocenter)", color='red', fontweight='bold')

    # Centroid (G)
    ax.scatter(*G, color='green', s=60, edgecolors='white', zorder=5)
    ax.text(G[0]+0.04, G[1], "G", color='green', fontweight='bold')

    # Nine-point (N)
    ax.scatter(*N, color='magenta', s=60, edgecolors='white', zorder=5)
    ax.text(N[0]+0.04, N[1], "N", color='magenta', fontweight='bold')

    # Circumcenter (O) and fixed vertices
    ax.scatter([O[0], A[0], B[0]], [O[1], A[1], B[1]], color='black', s=40, zorder=4)
    ax.text(O[0]-0.08, O[1]-0.08, "O", fontweight='bold')
    ax.text(A[0]+0.05, A[1], "A")
    ax.text(B[0]-0.08, B[1], "B")

    # Layout Aesthetics
    ax.set_aspect("equal")
    ax.set_xlim(-1.3, 1.3)
    ax.set_ylim(-1.3, 1.3)
    ax.grid(True, linestyle=':', alpha=0.4)
    ax.set_title("Kimberling Center Loci in the Thales Configuration", fontsize=14, pad=15)
    ax.legend(loc='lower left', frameon=True, shadow=True)
    
    plt.show()

display(interactive(plot_geometry, theta=(0.01, np.pi-0.01, 0.02)))
```

There are other centers that do not lie along the Euler line. For the parametrization:

$$
    X_n(\theta) = \lambda_n(\theta)(\cos\theta,\sin\theta)
$$

Let's consider:
- $X_1$: Incenter
- $X_2$: Centroid
- $X_3$: Circumcenter
- $X_4$: Orthocenter
- $X_5$: Nine-point center
- $X_6$: Symmedian

As
- Barycentric coordinates of the centroid is $(1 : 1 : 1)$
- Barycentric coordinates of the symmedian is $(a^2 : b^2 : c^2)$

Both points $X_1$ and $X_6$ lie outside of the Euler line. This is because $\lambda_1(\theta)$ and $\lambda_6(\theta)$ are not constants. How do these points vary as $\theta$ varies? Drag the slider below to see how the loci interact with $\theta$.

```python
def get_triangle_data(theta):
    # 1. Fixed Diameter AB, Variable Vertex C
    A = np.array([1, 0])
    B = np.array([-1, 0])
    C = np.array([np.cos(theta), np.sin(theta)])
    
    # 2. Side Lengths
    a = np.linalg.norm(C - B)
    b = np.linalg.norm(C - A)
    c = 2.0  # Constant diameter

    # 3. Compute Specific Centers
    # Euler Group
    O = np.array([0, 0])      # X3: Circumcenter (Fixed at Origin)
    H = C                     # X4: Orthocenter (Moves on unit circle)
    G = (A + B + C) / 3       # X2: Centroid (Moves on circle r=1/3)
    N = (O + H) / 2           # X5: Nine-point Center (Moves on circle r=1/2)
    
    # Off-Euler Group (Barycentric Calculation)
    # X1: Incenter = (aA + bB + cC) / (a + b + c)
    X1 = (a*A + b*B + c*C) / (a + b + c)
    
    # X6: Symmedian (Lemoine Point) = (a^2*A + b^2*B + c^2*C) / (a^2 + b^2 + c^2)
    X6 = (a**2*A + b**2*B + c**2*C) / (a**2 + b**2 + c**2)

    return {'X3': O, 'X4': H, 'X2': G, 'X5': N, 'X1': X1, 'X6': X6}, A, B, C

# --- Precompute Loci for Visual Trails ---
t_vals = np.linspace(0.01, np.pi-0.01, 400)
history = {k: [] for k in ['X4', 'X2', 'X5', 'X1', 'X6']}

for t in t_vals:
    centers, _, _, _ = get_triangle_data(t)
    for key in history:
        history[key].append(centers[key])

for key in history:
    history[key] = np.array(history[key])

def plot_thales_geometry(theta=1.0):
    centers, A, B, C = get_triangle_data(theta)
    
    fig, ax = plt.subplots(figsize=(9, 9))
    
    # Style Dictionary: (Color, LineStyle, Label)
    styles = {
        'X4': ('#e74c3c', '--', 'Orthocenter (X4)'),
        'X5': ('#9b59b6', '--', 'Nine-point (X5)'),
        'X2': ('#2ecc71', '--', 'Centroid (X2)'),
        'X1': ('#f39c12', '-',  'Incenter (X1)'),
        'X6': ('#3498db', '-',  'Symmedian (X6)')
    }

    # 1. Draw Loci Trails (The Tracks)
    for key, (clr, lst, lbl) in styles.items():
        ax.plot(history[key][:,0], history[key][:,1], color=clr, ls=lst, lw=4, alpha=0.2)

    # 2. Draw Euler Line (Connect X3 to X4)
    O, H = centers['X3'], centers['X4']
    ax.plot([O[0], H[0]], [O[1], H[1]], color='black', lw=1.5, ls=':', label='Euler Line')

    # 3. Draw Triangle
    ax.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], color='navy', lw=1.5, alpha=0.7)

    # 4. Draw Current Points
    for key, (clr, _, _) in styles.items():
        pos = centers[key]
        ax.scatter(*pos, color=clr, s=60, edgecolors='white', zorder=5)
        ax.text(pos[0]+0.02, pos[1]+0.02, key, color=clr, fontweight='bold', fontsize=10)
    
    # Circumcenter (Fixed)
    ax.scatter(*centers['X3'], color='black', s=40)
    ax.text(centers['X3'][0]-0.08, centers['X3'][1]-0.08, "X3 (O)", fontweight='bold')

    # 5. Final Touches
    ax.set_aspect("equal")
    ax.set_xlim(-1.2, 1.2); ax.set_ylim(-0.2, 1.1)
    ax.set_title("Loci of Kimberling Centers in Thales Configuration", fontsize=14)
    ax.legend(loc='upper left', fontsize='small', frameon=True)
    ax.grid(True, alpha=0.1)
    plt.show()

display(interactive(plot_thales_geometry, theta=(0.1, np.pi-0.1, 0.05)));
```

### Observations

- The centroid traces a circular locus when vertex $C$ moves on a circle.
- The nine-point center reflects the midpoint relation between circumcenter and orthocenter.
- Degenerate triangles occur when $A$, $B$ and $C$ become collinear; these cases are filtered using `NaN` checks.
